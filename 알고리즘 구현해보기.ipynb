#### score_x_y 구하기

```python
# 유저 x = 1814490, y=1681380
```

```python
len(sent_to_y)
```

```python
score_x_y = 0
# 집합 u에 대하여 반복
for i, u in enumerate(sent_to_y) :
    u_refrom = property_df[property_df['mem_no']==u]['view_list'].values[0]
    x_refrom = property_df[property_df['mem_no']==1817414]['view_list'].values[0]
    intersection = list(set(u_refrom) & set(x_refrom))
    union = list(set(u_refrom) | set(x_refrom))
    score_x_y += len(intersection) / len(union)
    print('score_x_y :',score_x_y)
```

#### score_y_x 구하기

```python
# x에게 메시지를 보낸 유저 집합 v는
sent_to_x = property_df[property_df['mem_no']==1814490]['viewed_list'].values[0]
sent_to_x
```

```python
score_y_x = 0
# 집합 v에 대하여 반복
for v in sent_to_x :
    v_refrom = property_df[property_df['mem_no']==v]['view_list'].values[0]
    y_refrom = property_df[property_df['mem_no']==1681380]['view_list'].values[0]
    intersection = list(set(v_refrom) & set(y_refrom))
    union = list(set(v_refrom) | set(y_refrom))
    score_y_x += len(intersection) / len(union)
    print('score_y_x :',score_y_x)
```

```python
# 조화평균 2ab / a+b
(2 * score_x_y * score_y_x) / (score_x_y + score_y_x)
```
